```python
from astroquery.jplhorizons import Horizons
import astropy.units as u
import numpy as np
```

```python
obj = Horizons(id='Ceres', location='568', epochs=2458133.33546)

# id is the object of interest
# location is the centre of the reference frame (568 refers to Maunakea (observatory), Hawaii)
# epochs is the time of interest but in Julian date (a continuous count of days and fractions since noon Universal Time on January 1, 4713 BC (on the Julian calendar))

print(obj)
```

# We can look across time ranges as well

```python
obj = Horizons(id='433', location='500@399',
               epochs={'start':'2010-01-01', 'stop':'2019-12-31',
                       'step':'1y'})

# in the above instance the time step is 1 year (you can change this). Look at documentation for available time steps.

eph = obj.ephemerides()
```

```python
print(eph.columns)
```

```python
eph["RA"]
```

```python
eph["datetime_jd"]
```

```python
# we can find at what index a certain time is

time_interest = 2456658.5

date_list = eph["datetime_jd"]

idx = np.where(date_list == time_interest)[0][0]

# we add the two [0] above to due to dimensionality of the np.where output
# have a play around to see what happens when you don't add them
# if time_interest was in date_list more than once np.where(date_list == time_interest)[0] would return multiple results
# np.where(date_list == time_interest)[0][0] would then grab the 1st (index 0) element of that list
# i.e. it would get the index when time_interest first appears

print(idx)
```

```python
# using idx we can get the corresponding RA value

RA_interest = eph["RA"][idx]
RA_interest
```

```python
# we can do this another way also
# as RA and date_list are the same length and have corresponding values

RA_list = eph["RA"]

RA_interest = RA_list[date_list == time_interest]

# date_list == time_interest effectively creates a mask that is then applied to RA_list

RA_interest
```

```python
# we can also filter the whole ephemaris (eph) table by the time of interest

eph[eph["datetime_jd"] == time_interest]
```

```python
# we can also filter to get times later than time_interest

filterer_date_list = date_list[date_list > time_interest]
filterer_date_list
```

```python
eph[eph["datetime_jd"] > time_interest]
```

```python
eph[eph["datetime_jd"] > time_interest]["DEC"]
```
